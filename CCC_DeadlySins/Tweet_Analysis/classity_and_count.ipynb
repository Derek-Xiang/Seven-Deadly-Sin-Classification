```python
def classifier_sins(cleaned_text):
    output = (None, None)
    count_all = 0
    # match_all = []
    for sin in sins:
        count = 0
        for word in sins_dict[sin]:
            match = re.findall(word,cleaned_text)
#             if match:
#                 match_all.append(match)
            num = len(match)
            count += num
 
        count_all += count 
        if output[0] == None:
            output = (sin, count)
        else:
            if count > output[1]:
                output = (sin, count)
    if count_all == 0:
        return None
    return output[0]
```

```python
def clean_text(text):
    # remove url | @ someone(name) | "\n"
    pattern = re.compile(r'https:[\S]*|@[\S]*|\\n|RT')
    step1 = re.sub(pattern, '', text)
#     # remove non-English character
#     pattern1 = re.compile(r'[^a-zA-Z\s]')
#     step2 = re.sub(pattern1, '', step1)
    return step1
```

```python
def process_row(row,coordinates,block,text,sin):
    dict_tweet = {}
    dict_tweet['_id'] = row['doc']['id_str']
    dict_tweet['coordinates'] = coordinates
    dict_tweet['block'] = block
    dict_tweet['created_date'] = str(row['doc']['created_at'][4:10])
    dict_tweet['created_weekday'] = str(row['doc']['created_at'][:3])
    dict_tweet['created_time'] = str(row['doc']['created_at'][11:19])
    dict_tweet['text'] = text
    dict_tweet['sin'] = sin
    return dict_tweet, dict_tweet['_id']
```

```python
def which_block(coordinate):
    p = Point(coordinate)
    for shape in suburbs['features']:
        poly=Polygon(shape['geometry']['coordinates'][0][0])
        if p.within(poly):
            return shape['properties']['vic_lga__3']
        else:
            continue
    # not in any block
    return False
```
